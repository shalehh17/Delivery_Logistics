```python
import joblib
import pandas as pd
from sklearn.compose import ColumnTransformer

# 1. Load dataset yang digunakan saat training
df = pd.read_csv('Delivery_Logistics_Clean.csv')
# 2. Definisikan ulang kolom fitur yang digunakan (sama persis dengan saat training)
required_cols = [
    'delivery_partner', 'package_type', 'vehicle_type', 'delivery_mode',
    'region', 'weather_condition', 'distance_km', 'package_weight_kg',
    'expected_time_hours', 'delivery_time_hours', 'delivery_cost', 'delivery_rating'
]
features = df[required_cols]

# 3. Load preprocessor yang lama (kerangka kosong)
preprocessor = joblib.load('models/preprocessor.pkl')

# 4. FIT DENGAN DATA (Ini langkah yang terlewatkan)
preprocessor.fit(features)

# 5. SIMPAN KEMBALI
joblib.dump(preprocessor, 'models/preprocessor.pkl')
print("Preprocessor berhasil di-fit dan disimpan!")
```
